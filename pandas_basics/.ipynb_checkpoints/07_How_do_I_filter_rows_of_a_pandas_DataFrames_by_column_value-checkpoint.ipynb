# 7. How do I filter rows of a pandas DataFrames by column value?

```python
import pandas as pd
movies = pd.read_csv('data/imdbratings.csv')
movies.head()
```

```python
# Filtering data details process
booleans = []
for length in movies['duration']:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)

print(f"lenght of booleans variable = {len(booleans)}")
is_long = pd.Series(booleans)

# passing a Series of booleans (True and False) to the DataFrame, we notice that it shows only 
# the values with True value, in this case the movies with duration more than 200 min.
movies[is_long].head()
```

```python
# Filtering rows by columns values
movies[movies['duration'] >= 200]
```

```python
# BONUS:
# select the 'genre' Series from the filtered DataFrame.
movies.loc[movies['duration'] >= 200, 'genre']
```
