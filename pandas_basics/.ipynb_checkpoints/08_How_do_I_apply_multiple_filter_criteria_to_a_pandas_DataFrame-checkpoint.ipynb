# 8. How do I apply multiple filter criteria to a pandas DataFrame?

```python
import pandas as pd
movies = pd.read_csv('data/imdbratings.csv')
movies.head()
```

```python
movies[movies['duration'] >= 200].head()
```

```python
# Applying filtering by multiple criteria, use '&', '|' instead of 'and', 'or'.
movies[(movies['duration'] >= 200) & (movies['genre'] == 'Drama')].head()
```

```python
movies[(movies['duration'] >= 200) | (movies['genre'] == 'Drama')].head()
```

```python
# BONUS
# If we have a long condition like this one we can use 'isin' methode.
movies[(movies['genre'] == 'Crime') | (movies['genre'] == 'Drama') | (movies['genre'] == 'Adventure')].head()
```

```python
movies[movies['genre'].isin(['Crime', 'Drama', 'Action'])].head()
```
