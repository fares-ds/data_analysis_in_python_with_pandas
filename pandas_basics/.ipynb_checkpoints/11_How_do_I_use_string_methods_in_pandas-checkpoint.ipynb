# 11. How do I use string methods in pandas?

```python
import pandas as pd
orders = pd.read_csv('data/chiporders.csv', index_col=0)
orders.head()
```

```python
# You can apply string methods on DataFrame columns value by using '.str' method.
# for example in this case we use '.str' to turn items names into lowercase letters.
orders['item_name'] = orders['item_name'].str.lower()
orders.head()
```

```python
# 'contains()': Check for a string if it contains certin character
orders[orders['item_name'].str.contains('chicken')].head()
```

```python
# BONUS
# replacing using regular expressions instead of .str.replace() method.
orders['choice_description'] = orders['choice_description'].str.replace('[', '').str.replace(']', '')
orders.head()
```

```python
# many pandas string methods support regular expressions (regex).
orders['choice_description'].str.replace('[\[\]]', '').head()
```
