# 29. How do I apply a function to a pandas Series or DataFrame?

```python
import pandas as pd
train = pd.read_csv('data/kaggletrain.csv')
train.head()
```

```python
# 'map()': Map values of Series according to input correspondence.

# Used for substituting each value in a Series with another value,
# that may be derived from a function, a ``dict`` or a :class:`Series`.
train['Sex_num'] = train.Sex.map({'female':0, 'male':1})
train.loc[0:4, ['Sex', 'Sex_num']]
```

```python
# 'apply()': Invoke function on values of Series.

# Can be ufunc (a NumPy function that applies to the entire Series)
# or a Python function that only works on single values.
train['Name_length'] = train.Name.apply(len)
train.loc[0:4, ['Name', 'Name_length']]
```

```python
# It is commun to use 'apply()' with a numpy function.
import numpy as np
train['Fare_ceil'] = train.Fare.apply(np.ceil)
train.loc[0:4, ['Fare', 'Fare_ceil']]
```

```python
train.Name.str.split(',').head()
```

```python
def get_element(my_list, position):
    return my_list[position]
```

```python
train.Name.str.split(',').apply(get_element, position=0).head()
```

```python
train.Name.str.split(',').apply(lambda my_list:my_list[0]).head()
```

```python
drinks = pd.read_csv('data/drinksbycountry.csv')
drinks.head()
```

```python
drinks.loc[:, 'beer_servings':'wine_servings'].apply(np.argmax, axis=1)
```

```python
# 'applymap()': Apply a function to a Dataframe elementwise.

# This method applies a function that accepts and returns a scalar to every element of a DataFrame.
drinks.loc[:, 'beer_servings':'wine_servings'] = drinks.loc[:, 'beer_servings':'wine_servings'].applymap(float)
drinks.head()
```
