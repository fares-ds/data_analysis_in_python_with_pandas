# 5. How do I remove columns from a pandas DataFrames?

```python
import pandas as pd
ufo = pd.read_csv('data/uforeports.csv', index_col=0)
ufo.head()
```

```python
# We use drop to delete columns or rows, 'columns': by passing 'name' and axis=1,
# 'rows': by passing row index and axis=0,
# or multiple columns and rows by passing a list of items to delete.
ufo.drop('Colors Reported', axis=1, inplace=True)
ufo.head()
```

```python
# Deleting multiple columns by passing list of columns name.
ufo.drop(['Location', 'Time'], axis=1).head()
```

```python
# Deleting multiple rows by passing list of rows name.
ufo.drop([0,4] , axis=0).head()
```
