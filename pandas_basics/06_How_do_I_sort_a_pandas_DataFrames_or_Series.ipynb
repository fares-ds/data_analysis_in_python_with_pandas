# 6. How do I sort a pandas DataFrames or Series?

```python
import pandas as pd
movies = pd.read_csv('data/imdbratings.csv')
movies.head()
```

```python
# 'sort_values()': is a DataFrame and Series method.
# Sort by the values along either axis.
# Here we sort Serie of titles.
movies['title'].sort_values(ascending=False).head()
```

```python
# Here we sort the movies DataFrame by title of the movies
movies.sort_values('title').head()
```

```python
# BONUS
# We can sort by multiple criteria by passing a list of columns you want to sort by.
movies.sort_values(['content_rating', 'title']).head()
```
